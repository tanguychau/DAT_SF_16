```python
import json
import urllib
import pandas as pd
import numpy as np
```

```python
your_key = "4WrTjTsFDGmHCLnYrAXeSLIjJYcfrG7j"
your_endpoint = "eb81bu78"

def getMovies(api_key=your_key,endpoint=your_endpoint):
    """
    Creates list of top 50 movies by gross box office
    sales for a with ratings and sales
    """  
    movies, ratings, sales, years = [], [], [], []
    #Remember to replace this link with the link to your specifc API
    url = "https://www.kimonolabs.com/api/{}?".format(endpoint) + \
            "apikey={}".format(api_key)
    
    data = json.load(urllib.urlopen(url))
    
    # Iterate through json object to collect data
    for n in xrange(data['count']):
        n_title = data['results']['collection1'][n]['title']['text']
        n_rating = float(data['results']['collection1'][n]['rating'])
        n_sales = float(data['results']['collection1'][n]['sales'][1:-1])
        n_year = int(data['results']['collection1'][n]['url'][-4:])
        movies.append(n_title)
        ratings.append(n_rating)
        sales.append(n_sales)
        years.append(n_year)
    data = pd.DataFrame({'movie':movies,'rating':ratings,'sales_in_M':sales,'year':years})    
    return data


movies = getMovies()
movies.head()
```

```python
import pandas.io.data
import datetime
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(figsize=(15,20))

# Count of ratings and sales
ax = fig.add_subplot(4,2,1)
ax.hist(movies.rating, bins=50)
ax.set_title('Distribution of Ratings', size=20)
ax.set_xlabel('count', size=16)

ax = fig.add_subplot(4,2,2)
ax.hist(movies.sales_in_M, bins=50)
ax.set_title('Distribution of Sales in $M', size=20)
ax.set_xlabel('count', size=16)

# Average rating and sales groupped by year
movies_byyear = movies.groupby('year').mean()
movies_rating_byyear = np.array(movies_byyear.values[:,0]).tolist()
movies_sales_byyear = np.array(movies_byyear.values[:,1]).tolist()

ax = fig.add_subplot(4,2,3)
ax.plot(list(set(movies.year)),movies_rating_byyear)
ax.set_title('Avg Rating by year', size=20)

ax = fig.add_subplot(4,2,4)
ax.plot(list(set(movies.year)),movies_sales_byyear)
ax.set_title('Avg Sales by year', size=20)

# Sales vs rating 
ax = fig.add_subplot(4,2,5)
ax.plot(movies.rating, movies.sales_in_M, 'o')
ax.set_title('Sales vs. Rating', size=20)

ax = fig.add_subplot(4,2,3)
ax.plot(list(set(movies.year)),movies_rating_byyear)
ax.set_title('Avg Sales vs. Rating', size=20)
```
